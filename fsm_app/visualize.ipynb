```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python

# read output of runs and accumlate the output of each thread
outputs = {}
with open('output.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    _ = next(csv_reader)
    for (thread, time) in csv_reader:
        if thread not in outputs:
            outputs[thread] = []
        outputs[thread].append(float(time))
        
```

```python
#calculate the average time for all threads
proc_output = {}
for (thread, timings) in outputs.items():
    avg = sum(timings) / len(timings)
    avg_approx = round(avg, 2)
    proc_output[thread] = avg_approx

print(proc_output)
```

```python
_, ax = plt.subplots()
ax.plot(proc_output.keys(), proc_output.values())

ax.set(xlabel='# of Threads', ylabel='Time (s)', title='Time of threads used for a grep search')
ax.grid()
plt.savefig("results.png")
plt.show()
```
